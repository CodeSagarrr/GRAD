## Chunking :

* Large documents ko chhote parts me todna chunking kehlata hai.

* Pure document ko ek saath embedding me dena ineffective hota hai.Isliye hum usko chhote meaningful tukdo me tod dete hai.

* Why chunking is critical :

    - Better semantic matching

    - Token limit safe rehti hai

    - Accurate retrieval

    - Context focused hota hai

* Chunk size → ek chunk me kitna text

* Overlap → adjacent chunks me kitna common text

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)


chunks = text_splitter.split_documents(document)
print(len(chunks))
```

## Enhancing recall :

* Recall ka matlab hai system kitni relevant information retrieve kar pa raha hai.

* Agar answer document me hai, lekin retrieval usko laa hi nahi raha, to recall low hai.

* Techniques to improve recall :

    - Query expansion : User query ko multiple queries me convert karna.

    - Multi vector retrieval : Ek document ke multiple embeddings banao , jaise summary level aur paragraph level.

    - Re ranking : Pehele zyada chunks retrieve karo, fir best chunks select karo.

```python
# Query expansion

queries = [
    "What is RAG?",
    "Explain retrieval augmented generation",
    "How RAG works in LLM"
]

# Multi vector retrieval ( Metadata filtering )

vector_store.similarity_search(
    query,
    filter={"source": "nlp_notes"}
)
```
